```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = {
    'Ticket': [
        'Internet not working',
        'Unable to login to account',
        'Payment failed during checkout',
        'Need refund for order',
        'Website loading very slowly',
        'Order not delivered',
        'App crashes frequently',
        'Password reset not working',
        'Received wrong product',
        'Server is down',
        'Billing issue on subscription',
        'Cannot access dashboard',
        'Delivery delayed',
        'Network connection problem',
        'Payment deducted twice',
        'Technical issue with app',
        'Refund not received',
        'Login error message',
        'Order tracking unavailable',
        'System outage affecting users'
    ],
    
    'Category': [
        'Technical',
        'Technical',
        'Billing',
        'Billing',
        'Technical',
        'Delivery',
        'Technical',
        'Technical',
        'Delivery',
        'Technical',
        'Billing',
        'Technical',
        'Delivery',
        'Technical',
        'Billing',
        'Technical',
        'Billing',
        'Technical',
        'Delivery',
        'Technical'
    ]
}

df = pd.DataFrame(data)

print(df.head())
```

```python
X = df['Ticket']
y = df['Category']

vectorizer = TfidfVectorizer()

X_vectorized = vectorizer.fit_transform(X)

print(X_vectorized.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_vectorized,
    y,
    test_size=0.2,
    random_state=42
)

model = MultinomialNB()

model.fit(X_train, y_train)

print("Model Trained Successfully")
```

```python
y_pred = model.predict(X_test)

print("Predicted Categories:")
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", round(accuracy, 2))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
priority_data = {
    'Ticket': [
        'Server is down',
        'Payment deducted twice',
        'Order delayed',
        'Unable to login',
        'Refund not received'
    ]
}

priority_df = pd.DataFrame(priority_data)

def assign_priority(ticket):
    ticket = ticket.lower()

    if 'down' in ticket or 'payment deducted' in ticket:
        return 'High'
    elif 'login' in ticket or 'refund' in ticket:
        return 'Medium'
    else:
        return 'Low'

priority_df['Priority'] = priority_df['Ticket'].apply(assign_priority)

print(priority_df)
```

```python
new_tickets = [
    "Payment issue in subscription",
    "Website not loading",
    "Order not received"
]

new_vectors = vectorizer.transform(new_tickets)

predictions = model.predict(new_vectors)

for ticket, category in zip(new_tickets, predictions):
    print(f"Ticket: {ticket}")
    print(f"Predicted Category: {category}")
    print("-" * 40)
```
